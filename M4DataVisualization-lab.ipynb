## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns

QUERY = """
    SELECT  Respondent, ConvertedComp
    FROM master
    """
df_hist = pd.read_sql_query(QUERY, conn)
df_hist['ConvertedComp'].plot(kind='hist', figsize=(8, 6), alpha=1.0)
plt.title('Converted Compensation')
plt.xlabel('Compensation in USD')
plt.ylabel('Respondents')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = """ SELECT Respondent, Age
            FROM master
        """
df_age = pd.read_sql_query(QUERY, conn)
df_age['Age'].plot(kind='box')
plt.title('Age of Respondents')
plt.ylabel('Respondent')

plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """ SELECT WorkWeekHrs, Age
            FROM master
        """
df_age_hours = pd.read_sql_query(QUERY, conn)
df_age_hours.plot(kind='scatter', x='Age', y='WorkWeekHrs', figsize=(10,6), color='darkblue')
plt.title('Age vs Hours Worked during Week')
plt.xlabel('Age')
plt.ylabel('Hours Worked/Week')

plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
QUERY = """ SELECT CodeRevHrs, WorkWeekHrs, Age
            FROM master
        """
df_bubble = pd.read_sql_query(QUERY, conn)

sns.scatterplot(data=df_bubble, x='WorkWeekHrs', y='CodeRevHrs', size='Age', hue='Age', color='green', alpha=1.0, sizes=(100, 8))
plt.title('WorkWeekHrs, CodeRevHrs vs Age')
plt.xlabel('Work Week Hours')
plt.ylabel('CodeRevHrs')
plt.show()
```

## Visualizing composition of data


### Pie Charts


```python
#Create a pie chart of the top 5 programming languages that respondents wish to lean next year. Label the pie chart with database names. Display percentages of each database on the pie chart.

QUERY = """
        SELECT LanguageDesireNextYear, count(*) as count
        FROM LanguageDesireNextYear
        GROUP BY LanguageDesireNextYear
        ORDER BY count(LanguageDesireNextYear)
        DESC
        LIMIT 5
        """

df_lang = pd.read_sql_query(QUERY, conn)
df_lang.set_index('LanguageDesireNextYear', inplace=True)

colors_list = ['red', 'green', 'blue', 'orange', 'purple']
df_lang['count'].plot(kind='pie', figsize=(10,6), autopct='%1.10f%%', labels=None, startangle=90, colors=colors_list, shadow=True, pctdistance=1.12)


plt.legend(labels=df_lang.index, loc='upper right')
plt.title('Top 5 Programmin Languages Respondents Wish To Learn') 
plt.axis('equal')
plt.show()
        
```

```python
#How many respondents indicated they currently work with 'SQL'.

QUERY = """
        SELECT LanguageWorkedWith, count(*) as count
        FROM LanguageWorkedWith
        GROUP BY LanguageWorkedWith
        ORDER BY count(LanguageWorkedWith)
        DESC
        LIMIT 5        
        """

df_lang_worked_with = pd.read_sql_query(QUERY, conn)
df_lang_worked_with.set_index('LanguageWorkedWith', inplace=True)
print(df_lang_worked_with)    
 
df_lang_worked_with['count'].plot(kind='pie', figsize=(10,6), autopct='%1.1f%%', labels=None, startangle=90, colors=colors_list, shadow=True, pctdistance=1.12, explode=(0, 0, 0.1, 0, 0)  )

plt.legend(labels=df_lang_worked_with.index, loc='upper right')
plt.title('Top 5 Programming Languages Respondents Work With') 
plt.axis('equal')
plt.show()
```

```python
#Show how many respondents indicated that they work on 'MySQL' only

QUERY = """
        SELECT Respondent, DatabaseWorkedWith, COUNT(*) as count
        FROM DatabaseWorkedWith
        GROUP BY Respondent
        HAVING count = 1
            AND DatabaseWorkedWith = 'MySQL'
                                    
       """
df_MySQL = pd.read_sql_query(QUERY, conn)
print('The number of Respondents working with MySQL only is:', df_MySQL.shape[0])
```

```python
#Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.

# your code goes here
QUERY = """ 
SELECT DatabaseDesireNextYear, count(*) as count
from DatabaseDesireNextYear 
GROUP BY DatabaseDesireNextYear
ORDER BY count(DatabaseDesireNextYear) 
DESC
LIMIT 5

""" 

df=pd.read_sql_query(QUERY,conn)  
df.set_index('DatabaseDesireNextYear', inplace=True) 
colors_list=['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'pink']

df['count'].plot(kind='pie', figsize=(10,6), autopct='%1.1f%%', labels=None, startangle=90, colors=colors_list, shadow=True, pctdistance=1.12)  

plt.legend(labels=df.index, loc='upper right')
plt.title('Top 5 Databases Respondents Wish To Learn') 
plt.axis('equal')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

QUERY = """ SELECT WorkWeekHrs, CodeRevHrs, Age
            FROM master
            WHERE Age BETWEEN 30 AND 35
        """
df_stacked = pd.read_sql_query(QUERY, conn)
df_stacked_2 = df_stacked.groupby('Age').median()

df_stacked_2.plot(kind='bar', figsize=(15, 5), stacked=True)
plt.xlabel('Age')
plt.ylabel('Hours')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title('WorkWeekHrs and CodeRevHrs Vs Age')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """ SELECT ConvertedComp, Age
            FROM master
            WHERE Age BETWEEN 45 AND 60
        """
df_line = pd.read_sql_query(QUERY, conn)

df_line = df_line.loc[(df_line['Age'] >= 45) & (df_line['Age'] <= 60)].reset_index().drop(columns=['index'])
df_line.dropna(inplace=True)
   
df_line = df_line.groupby('Age', axis=0).median()
    
df_line.plot(kind='line', figsize=(15, 5))
plt.xlabel('Age')
plt.ylabel('Compensation')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title('Converted Compensation Vs Age')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """
        SELECT MainBranch, COUNT(*) as MainBranch
        FROM master
        GROUP BY MainBranch
        """
df_main = pd.read_sql_query(QUERY, conn)
df_main.plot(kind='barh', figsize=(8, 6), alpha=1.0, color='teal')
plt.title('Where Respondents Work')
plt.xlabel('Number')
plt.ylabel('Location 0: Not at the Main Branch\n' + 'Location 1: Main Branch')
plt.show()

```
