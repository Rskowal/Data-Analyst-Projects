```python
# Get the value counts
df_counts = df['YearsCodePro'].value_counts().reset_index()
df_counts.columns = ['YearsCodePro', 'Count']

# Replace text values with numerical equivalents
df_counts['YearsCodePro'] = df_counts['YearsCodePro'].replace({'Less than 1 year': 0, 'More than 50 years': 51})

# Convert the column to numeric type
df_counts['YearsCodePro'] = pd.to_numeric(df_counts['YearsCodePro'])

# Sort the DataFrame by 'YearsCodePro'
df_counts = df_counts.sort_values(by='YearsCodePro')

# Plotting the line graph
plt.figure(figsize=(10, 6))
plt.plot(df_counts['YearsCodePro'], df_counts['Count'], marker='o', linestyle='-', color='skyblue')

# Adding titles and labels
plt.title('Distribution of Years of Professional Coding Experience')
plt.xlabel('Years of Professional Coding')
plt.ylabel('Count')
plt.show()
```

```python
import numpy as np
```

```python
# Replace "Less than 1 year" and "More than 50 years" with 0 and 51, respectively
df['YearsCodePro'] = df['YearsCodePro'].replace({
    np.nan: 0,
    'Less than 1 year': 0, 
    'More than 50 years': 51
}).astype(int)

# Fill any missing values (NaN) with a suitable value, e.g., 0
df['YearsCodePro'] = df['YearsCodePro'].fillna(0).astype(int)

# Convert the 'Age' column to categorical numerical values
age_mapping = { 'Under 18 years old': 1,
    '18-24 years old': 2, '25-34 years old': 3, 
    '35-44 years old': 4, '45-54 years old': 5, 
    '55-64 years old': 6, '65 years or older': 7
}
df['AgeNum'] = df['Age'].map(age_mapping)

# Calculate the mean YearsCodePro for each age group
mean_years_code_pro = df.groupby('AgeNum')['YearsCodePro'].mean().reset_index()

# Plotting the data
plt.plot(mean_years_code_pro['AgeNum'], mean_years_code_pro['YearsCodePro'], marker='o')
plt.xlabel('Age')
plt.ylabel('Mean Years of Professional Coding')
plt.title('Mean Years of Professional Coding vs. Age')
plt.xticks(list(age_mapping.values()), list(age_mapping.keys()), rotation=45, ha='right')
plt.show()
```

```python
# Create a DataFrame with age distribution by gender
df_gender_age = df.groupby(['Age', 'Gender']).size().unstack().fillna(0)

# Plot the results in a line chart
plt.figure(figsize=(10, 6))
plt.plot(df_gender_age.index, df_gender_age['Male'], label='Male')
plt.plot(df_gender_age.index, df_gender_age['Female'], label='Female')

# Adding titles and labels
plt.title('Age Distribution by Gender')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend()

# Show the plot
plt.show()
```
