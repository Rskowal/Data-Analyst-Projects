### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
# Split the languages into separate rows for both columns
df_have = df.assign(LanguageHaveWorkedWith=df['LanguageHaveWorkedWith'].str.split(';')).explode('LanguageHaveWorkedWith')
df_want = df.assign(LanguageWantToWorkWith=df['LanguageWantToWorkWith'].str.split(';')).explode('LanguageWantToWorkWith')

# Calculate frequency of each programming language in both columns
have_counts = df_have['LanguageHaveWorkedWith'].value_counts().reset_index()
have_counts.columns = ['Language', 'HaveWorkedWith']

want_counts = df_want['LanguageWantToWorkWith'].value_counts().reset_index()
want_counts.columns = ['Language', 'WantToWorkWith']

# Merge the counts DataFrames
merged_counts = pd.merge(have_counts, want_counts, on='Language', how='outer').fillna(0)
merged_counts.set_index('Language', inplace=True)
print("\nMerged Counts DataFrame:\n", merged_counts)

# Visualize with a grouped bar chart
merged_counts.plot(kind='bar', figsize=(12, 8))
plt.title('Comparison of Programming Languages Have Worked With vs Want to Work With')
plt.xlabel('Programming Language')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
## Write your code here
country_remote_work = df.groupby(['Country', 'RemoteWork']).size().reset_index(name='Count')
pivot_table = country_remote_work.pivot(index='RemoteWork', columns='Country', values='Count').fillna(0)

plt.figure(figsize=(15, 10))
sns.heatmap(pivot_table, cmap='coolwarm', annot=True, fmt='g')
plt.title('Distribution of Remote Work Preferences by Country')
plt.xlabel('Country')
plt.ylabel('Remote Work')
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
!pip install scipy
import scipy
```

```python
from scipy.stats import pearsonr, spearmanr
```

```python
# Write your code here

def handle_special_values(years):
    if years == 'Less than 1 year':
        return 0
    elif years == 'More than 50 years':
        return 51
    else:
        return int(years)


df['YearsCodePro'] = df['YearsCodePro'].apply(handle_special_values)
print(f"\nDataFrame after handling special values in YearsCodePro:", df['YearsCodePro'].head(25))
pearson_corr, _ = pearsonr(df['YearsCodePro'], df['JobSat']) 
print(f"\nPearson correlation coefficient: {pearson_corr}")
spearman_corr, _ = spearmanr(df['YearsCodePro'], df['JobSat']) 
print(f"Spearman correlation coefficient: {spearman_corr}")
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
## Write your code here
crosstab = pd.crosstab(df['Employment'], df['EdLevel'])

crosstab.plot(kind='bar', stacked=True, figsize=(12, 8), colormap='coolwarm')
plt.title('Cross-tabulation of Employment Status by Education Level')
plt.xlabel('Employment Status')
plt.ylabel('Count')
plt.legend(title='Education Level')
plt.show()
```
