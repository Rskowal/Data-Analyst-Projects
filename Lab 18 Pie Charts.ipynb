### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
##Write your code here
# Step 1: Expand 'DatabaseWantToWorkWith' into multiple rows
df_expanded = df.assign(DatabaseWantToWorkWith=df['DatabaseWantToWorkWith'].str.split(';')).explode('DatabaseWantToWorkWith')

# Step 2: Calculate frequencies
frequency = df_expanded['DatabaseWantToWorkWith'].value_counts().reset_index()
frequency.columns = ['DatabaseWantToWorkWith', 'Frequency']

# Step 3: Select the top 5 databases
top_5_databases = frequency.head(5)

# Step 4: Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_5_databases['Frequency'], labels=top_5_databases['DatabaseWantToWorkWith'], autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Most-Desired Databases')
plt.show()
```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.
