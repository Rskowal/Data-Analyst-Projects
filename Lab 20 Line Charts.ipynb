### Task 1: Trends in Compensation Over Age Groups


##### 1. Line Chart of Median `ConvertedCompYearly` by Age Group


- Track how the median yearly compensation (ConvertedCompYearly) changes across different age groups.

- Use a line chart to visualize these trends.


```python
# Filter out "Prefer not to say" values
df = df[(df['Age'] != 'Prefer not to say') | (df['Age'] != 'Under 19 years old')]

# Group by Age and calculate the mean ConvertedCompYearly
mean_compensation_by_age = df.groupby('Age')['ConvertedCompYearly'].mean()

# Define the custom sorting order
age_order = [
    'Under 18 years', 
    '18-24 years', 
    '25-34 years', 
    '35-44 years', 
    '45-54 years', 
    '55-64 years', 
    '65 years or older'
]

# Sort the Series by the custom order
mean_compensation_by_age = mean_compensation_by_age   #.reindex(age_order).dropna()

# Print the mean compensation by age group
print(mean_compensation_by_age)

# Plot the data
mean_compensation_by_age.plot(kind='line', marker='o', figsize=(10, 6))
plt.title('Mean Yearly Compensation by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Mean Yearly Compensation')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.show()
```

##### 2. Line Chart of Median `ConvertedCompYearly` for Ages 25 to 45


For a closer look, plot a line chart focusing on the median compensation for respondents between ages 25 and 45.


```python
AgeRange = df['Age'].unique().tolist()
print(AgeRange)
age_counts = df['Age'].value_counts()

data = { 'AgeRange': ['Under 18 years old', '18-24 years old', '25-34 years old', '35-44 years old', '45-54 years old', '55-64 years old', '65 years old', 'Prefer not to say'],
         'Counts': [2568, 14098, 23911, 14942, 6249, 2575, 772, 322]}
df1 = pd.DataFrame(data)
# Function to generate a random age within the specified range
def generate_random_age(age_range): 
    if age_range == 'Under 18 years old': 
        return np.random.randint(15, 18)
    elif age_range == '18-24 years old':
        return np.random.randint(18, 25)
    elif age_range == '25-34 years old':
        return np.random.randint(25, 35)
    elif age_range == '35-44 years old':
        return np.random.randint(35, 45)
    elif age_range == '45-54 years old':
        return np.random.randint(45, 55)
    elif age_range == '55-64 years old':
        return np.random.randint(55, 65)
    elif age_range == '65 years or older':
        return np.random.randint(65, 101)    # Assuming 100 as the upper limit for simplicity
    elif age_range == 'Prefer not to say':
        return None
    else: 
        start_age, end_age = map(int, age_range.split('-'))
        return np.random.randint(start_age, end_age + 1)
        

# Apply the function to the AgeRange column to generate random ages
df['RandomAge'] = df['Age'].apply(generate_random_age)

# Display the DataFrame with the new RandomAge column
print(df['RandomAge'])
```

```python
# Filter data for ages 25 to 45
df_filtered = df[(df['RandomAge'] >= 25) & (df['RandomAge'] <= 45)]
# Calculate the median compensation for each age
median_compensation_by_age = df_filtered.groupby('RandomAge')['ConvertedCompYearly'].median() 
# Print the median compensation by individual age
print(median_compensation_by_age)
# Plot the data
median_compensation_by_age.plot(kind='line', marker='o', figsize=(10, 6))
plt.title('Median Yearly Compensation for Ages 25 to 45')
plt.xlabel('Age') 
plt.ylabel('Median Yearly Compensation')
plt.grid(True)
plt.show()
```
